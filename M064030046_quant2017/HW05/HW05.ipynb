# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import pandas_datareader.data as web
from datetime import datetime
import pandas as pd
```

```python
def get_stock_data(stockname,startdate):
    url = "https://finance.google.com/finance/historical?q={}&startdate={}&output=csv".format(stockname,startdate)
    stockprice = pd.read_csv(url)
    print(stockprice)
```

```python
get_stock_data('AAPL',"2016-01-01")
```
